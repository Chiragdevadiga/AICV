```python
import cv2
```

```python
#reading the image
img=cv2.imread('cricket.jpg')
```

```python
# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Load the Haar cascade for face detection
haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Apply face detection on the grayscale image
faces_rect = haar_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=9)

# Iterate through rectangles of detected faces
for (x, y, w, h) in faces_rect:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)

# Display the image with detected faces
cv2.imshow('Detected faces', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Face and eye together

```python
import cv2
```

```python
#reading the image
img=cv2.imread('pratheek.jpeg')
```

```python
import cv2



# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Load the Haar cascade for face and eye detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Apply face detection on the grayscale image
face_rect = face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=9)

# Iterate through rectangles of detected faces
for (x, y, w, h) in face_rect:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)

# Apply eye detection on the grayscale image
eye_rect = eye_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=9)

# Iterate through rectangles of detected eyes
for (x, y, w, h) in eye_rect:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)  # Change color to (255, 0, 0) for eyes

# Display the image with detected faces and eyes
cv2.imshow('Detected faces and eyes', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
